```r
#chamar os pacotes:
library(tidyverse)
library(ggplot2)
```

```r
arq <- read.table('cancer.txt', header=TRUE) #lendo o arquivo.
g <- ggplot(iris)
```



---


**1.** Faça o histograma das variáveis LDH, ALB e N. Use a descrição da coluna como título de cada um. Salve o gráfico no formato pdf.


---



```r
#Variável LDH
ggplot(data = arq, aes(x = LDH)) +
  labs(title = "Enzima, Lactate Dehydrogenase",
       x = "Níveis de Enzima, Lactate Dehydrogenase",
       y =  "Quantidade de Resultados") +
  geom_histogram(color = "white",fill = "red", bins = 30) +
  theme_classic(base_size = 18)
  ggsave("graficLDH.pdf", units = "cm", width = 20, height = 20)

#Variável ALB
ggplot(data = arq, aes(x = ALB)) +
  labs(title = "Albumina",
       x = "Níveis de Albumina",
       y = "Quantidade de Resultados")+
  geom_histogram(color = "white", fill = "lightgreen", bins = 30) +
  theme_classic(base_size = 18)
  ggsave("graficALB.pdf", units = "cm", width = 20, height = 20)

#Variável N
ggplot(data = arq, aes(x = N)) +
  labs(title = "Nitrogênio na uréia", 
       x = "Níveis de Nitrogênio na uréia",
       y = "Quantidade de Resultados") +
  geom_histogram(color = "white", fill = "orange", bins=30) +
  theme_classic(base_size = 18)
  ggsave("graficN.pdf", units = "cm", width = 20, height = 20)
```

**3.** Médicos afirmam que o grupo dos falso-positivos é mais jovem do que o dos falso negativos. Compare as idades destes dois grupos usando gráfico de barras. Salve o gráfico no

**4.** Compare a glicose dos pacientes falso positivos e dos falso negativos. Para isso, use o
gráfico boxplot.


---



```r
#Níveis de glicose
ggplot(arq, aes(x = Grupo, y = GL, group = Grupo)) +
  geom_boxplot(color = "black", fill = "red") +
  labs(title = "Glicose vs Resultados") +
  xlab("1-Falso-negativo, 2-Negativo, 3-Positivo, 4-Falso-positivo") 
```



---

**5.** Faça um gráfico de dispersão comparando Albumina e Glicose dos pacientes.


---



```r
#Comparando Albumina e Glicose dos pacientes
ggplot(arq, aes(x = ALB, y = GL)) +
  geom_point(color = "violet") +
  theme_bw() +
  labs(title = "Comparando Albumina e Glicose dos pacientes",
        x = "Glicose",
        y = "Albumina") 
```



---

**6.** Repita o gráfico usando uma linha de tendência.


---



```r
#Comparando Albumina e Glicose dos pacientes com linha de tendência
ggplot(arq, aes(x = ALB, y = GL)) +
  geom_point(color = "green") +
  geom_smooth() +
  theme_bw() +
  labs(title = "Comparando Albumina e Glicose dos pacientes",
        x = "Glicose",
        y = "Albumina") 
```
